```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
import os
import json
import numpy as np

x = [1, 2, 3, 4, 5]
x_axis = [1, 2, 4, 8, 16]
used_memory_usm=[0, 1, 2, 3, 4, 5]  # fill in the this manually
memory_usage_usm = []

used_memory_nousm = [0,2,6,14,30,62]
memory_usage_nousm = []

for i in range(0, 5):
    memory_usage_usm.append(used_memory_usm[i + 1] - used_memory_usm[i])
    memory_usage_nousm.append(used_memory_nousm[i + 1] - used_memory_nousm[i]) 

plt.xticks(x, x_axis)

plt.plot(x, memory_usage_usm, 'o-', label="With USM")

plt.plot(x, memory_usage_nousm, 'o-', label="Without USM")

# plt.title("System Memory Usage")
plt.xlabel('Number of Concurrent Containers')
plt.ylabel('Memory Usage (MB)')
# plt.ylim(ymin=0)
plt.legend()
# plt.savefig(os.path.join(usm_data_dir, machine, "pictures", 'function_time.jpg'))
```
