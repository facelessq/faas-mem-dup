```python
# on einstein vm
import pandas as pd
import matplotlib.pyplot as plt
import sys
import os
import json
import numpy as np

x = [1, 2, 3, 4, 5]
x_axis = [1, 2, 4, 8, 16]
used_memory_usm=[474340, 786408, 1042492, 1556800, 2621212, 4701828]  # fill in the this manually
memory_usage_usm = []

used_memory_nousm = [515536,841168,1172096,1847696,3201996,5936268]  # fill in this
memory_usage_nousm = []

for i in range(1, 6):
    memory_usage_usm.append((used_memory_usm[i] - used_memory_usm[0]) / 1024)
    memory_usage_nousm.append((used_memory_nousm[i] - used_memory_nousm[0]) / 1024)

plt.xticks(x, x_axis)

plt.plot(x, memory_usage_usm, 'o-', label="With USM")

plt.plot(x, memory_usage_nousm, 'o-', label="Without USM")

plt.grid(axis='y')
plt.title("System Memory Usage")
plt.xlabel('Number of Concurrent Containers')
plt.ylabel('Memory Usage (MB)')
plt.rcParams['figure.dpi'] = 300
# plt.ylim(ymin=0)
plt.legend()
# plt.savefig(os.path.join(usm_data_dir, machine, "pictures", 'function_time.jpg'))
```
